```python
# Import required libraries
import pandas as pd
import numpy as np
import gc
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.cluster import KMeans
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.metrics import (
    accuracy_score, precision_score, recall_score, f1_score,
    silhouette_score, confusion_matrix
)

# Set random seed for reproducibility
np.random.seed(42)

# 1. Data Loading and Initial Inspection
print("1. DATA LOADING AND INITIAL INSPECTION")
print("=" * 50)

file_path = r"D:\Capstone projects\Disease Outbreak\disease_outbreak_data.csv"  # Adjust with your actual file name
df = pd.read_csv(file_path)

# Display basic information
print("\nDataset Info:")
print("-" * 30)
df.info()

print("\nFirst few rows:")
print("-" * 30)
display(df.head())

print("\nBasic statistics:")
print("-" * 30)
display(df.describe())

# 2. Data Cleaning and Preprocessing
print("\n2. DATA CLEANING AND PREPROCESSING")
print("=" * 50)

# Check for missing values
print("\nMissing values:")
print("-" * 30)
missing_values = df.isnull().sum()
print(missing_values[missing_values > 0])

# Handle missing values
print("\nHandling missing values...")
for col in ['vaccination_rate', 'healthcare_accessibility_score']:
    median_value = df[col].median()
    df[col].fillna(median_value, inplace=True)

# Check for outliers using IQR method
def detect_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)][column]
    return len(outliers)

# Check outliers in numerical columns
print("\nOutlier Detection:")
print("-" * 30)
numerical_cols = df.select_dtypes(include=[np.number]).columns
for col in numerical_cols:
    n_outliers = detect_outliers(df, col)
    if n_outliers > 0:
        print(f"{col}: {n_outliers} outliers detected")

# 3. Feature Engineering
print("\n3. FEATURE ENGINEERING")
print("=" * 50)

# Create new features
df['population_health_index'] = df['population_density'] * (1 - df['healthcare_accessibility_score'])
df['travel_risk_score'] = df['international_travel_freq'] * df['public_transport_usage'] / 100
df['environmental_risk'] = df['air_quality_index'] * df['avg_humidity'] / 100
df['healthcare_coverage'] = df['vaccination_rate'] * df['healthcare_accessibility_score']
df['outbreak_severity'] = df['disease_incidents_last_year'] * df['avg_outbreak_duration']

# 4. Initial Exploratory Data Analysis (EDA)
print("\n4. EXPLORATORY DATA ANALYSIS")
print("=" * 50)

# Create correlation matrix
plt.figure(figsize=(12, 8))
correlation_matrix = df.select_dtypes(include=[np.number]).corr()
sns.heatmap(correlation_matrix, cmap='coolwarm', center=0, annot=False)
plt.title('Correlation Matrix of Numerical Features')
plt.tight_layout()
plt.show()

# Risk distribution by region
plt.figure(figsize=(10, 6))
sns.boxplot(x='region_type', y='risk_score', data=df)
plt.title('Risk Score Distribution by Region Type')
plt.tight_layout()
plt.show()

# Distribution of key engineered features
engineered_features = ['population_health_index', 'travel_risk_score', 
                      'environmental_risk', 'healthcare_coverage', 
                      'outbreak_severity']

plt.figure(figsize=(15, 10))
for i, feature in enumerate(engineered_features, 1):
    plt.subplot(3, 2, i)
    sns.histplot(df[feature], kde=True)
    plt.title(f'{feature.replace("_", " ").title()} Distribution')
plt.tight_layout()
plt.show()

# Print summary statistics of engineered features
print("\nEngineered Features Statistics:")
print("-" * 30)
display(df[engineered_features].describe())

# Memory cleanup
plt.close('all')
gc.collect()
```

```python
from sklearn.metrics import (
    accuracy_score, precision_score, recall_score, f1_score,
    silhouette_score, roc_auc_score, roc_curve
)
# 5. Data Preparation for Modeling
print("5. DATA PREPARATION FOR MODELING")
print("=" * 50)

# Select features for modeling
model_features = [
    'population_density', 'vaccination_rate', 'healthcare_accessibility_score',
    'air_quality_index', 'avg_humidity', 'international_travel_freq',
    'public_transport_usage', 'population_health_index', 'travel_risk_score',
    'environmental_risk', 'healthcare_coverage', 'outbreak_severity'
]

# Prepare feature matrix
X = df[model_features]

# Scale features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_scaled = pd.DataFrame(X_scaled, columns=model_features)

# Encode target variable
le = LabelEncoder()
y = le.fit_transform(df['outbreak_risk_category'])

# 6. Clustering Analysis
print("\n6. CLUSTERING ANALYSIS")
print("=" * 50)

# Find optimal number of clusters
silhouette_scores = []
inertias = []
K = range(2, 6)

for k in K:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X_scaled)
    silhouette_scores.append(silhouette_score(X_scaled, kmeans.labels_))
    inertias.append(kmeans.inertia_)
    print(f"k={k}: Silhouette Score = {silhouette_scores[-1]:.3f}")

# Plot elbow curve and silhouette scores
plt.figure(figsize=(15, 5))
plt.subplot(1, 3, 1)
plt.plot(K, inertias, 'bx-')
plt.xlabel('k')
plt.ylabel('Inertia')
plt.title('Elbow Method for Optimal k')

plt.subplot(1, 3, 2)
plt.plot(K, silhouette_scores, 'rx-')
plt.xlabel('k')
plt.ylabel('Silhouette Score')
plt.title('Silhouette Score vs k')

# Apply optimal clustering
optimal_k = K[np.argmax(silhouette_scores)]
kmeans = KMeans(n_clusters=optimal_k, random_state=42)
df['cluster'] = kmeans.fit_predict(X_scaled)

# 7. Classification Modeling
print("\n7. CLASSIFICATION MODELING")
print("=" * 50)

# Split the data
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# Train Random Forest model
rf = RandomForestClassifier(
    n_estimators=100,
    max_depth=10,
    min_samples_split=5,
    random_state=42
)

# Perform cross-validation
cv_scores = cross_val_score(rf, X_train, y_train, cv=5)
print("\nCross-validation scores:", cv_scores)
print(f"Average CV Score: {cv_scores.mean():.3f} (+/- {cv_scores.std() * 2:.3f})")

# Train final model
rf.fit(X_train, y_train)
y_pred = rf.predict(X_test)
y_prob = rf.predict_proba(X_test)

# Calculate ROC-AUC score
roc_auc = roc_auc_score(y_test, y_prob, multi_class='ovr', average='weighted')

# Plot ROC curves
plt.subplot(1, 3, 3)
for i in range(len(le.classes_)):
    fpr, tpr, _ = roc_curve(y_test == i, y_prob[:, i])
    plt.plot(fpr, tpr, label=f'{le.classes_[i]} (AUC = {roc_auc:.2f})')

plt.plot([0, 1], [0, 1], 'k--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curves by Risk Category')
plt.legend(loc='lower right', fontsize='small')
plt.grid(True)
plt.tight_layout()
plt.show()

# Print complete classification metrics
print("\nClassification Metrics:")
print("-" * 30)
print(f"Accuracy: {accuracy_score(y_test, y_pred):.3f}")
print(f"Precision: {precision_score(y_test, y_pred, average='weighted'):.3f}")
print(f"Recall: {recall_score(y_test, y_pred, average='weighted'):.3f}")
print(f"F1 Score: {f1_score(y_test, y_pred, average='weighted'):.3f}")
print(f"ROC-AUC Score: {roc_auc:.3f}")

# Feature importance analysis
feature_importance = pd.DataFrame({
    'feature': model_features,
    'importance': rf.feature_importances_
}).sort_values('importance', ascending=False)

plt.figure(figsize=(10, 6))
sns.barplot(x='importance', y='feature', data=feature_importance)
plt.title('Feature Importance in Risk Prediction')
plt.tight_layout()
plt.show()

# 8. Cluster Analysis
print("\n8. CLUSTER ANALYSIS")
print("=" * 50)

# Analyze clusters
cluster_stats = df.groupby('cluster').agg({
    'risk_score': 'mean',
    'population_density': 'mean',
    'healthcare_accessibility_score': 'mean',
    'vaccination_rate': 'mean',
    'international_travel_freq': 'mean',
    'outbreak_risk_category': lambda x: x.value_counts().index[0]
}).round(2)

print("\nCluster Characteristics:")
print(cluster_stats)

# Clean up
plt.close('all')
gc.collect()
```

```python
# 9. FINAL COMPREHENSIVE ANALYSIS
print("COMPREHENSIVE HEALTHCARE RISK ANALYSIS")
print("=" * 50)

# Previous code remains the same until Model Performance by Region section

print("\n3. Model Performance by Region:")
print("-" * 40)

# Calculate predictions and performance metrics for each region
for region in df['region_type'].unique():
    region_mask = df['region_type'] == region
    region_y = le.fit_transform(df[region_mask]['outbreak_risk_category'])
    region_X = X_scaled[region_mask]
    
    # Split data for this region
    X_train_reg, X_test_reg, y_train_reg, y_test_reg = train_test_split(
        region_X, region_y, test_size=0.2, random_state=42
    )
    
    # Train and evaluate model for this region
    rf_reg = RandomForestClassifier(
        n_estimators=100,
        max_depth=10,
        min_samples_split=5,
        random_state=42
    )
    rf_reg.fit(X_train_reg, y_train_reg)
    
    # Calculate metrics
    y_pred_reg = rf_reg.predict(X_test_reg)
    accuracy_reg = accuracy_score(y_test_reg, y_pred_reg)
    
    print(f"\n{region} Areas:")
    print(f"â€¢ Accuracy: {accuracy_reg:.3f}")
    print(f"â€¢ Sample Size: {len(region_y)}")
    print(f"â€¢ Risk Categories Distribution:")
    risk_dist = df[region_mask]['outbreak_risk_category'].value_counts()
    for category, count in risk_dist.items():
        print(f"  - {category}: {count} ({count/len(region_y)*100:.1f}%)")

print("\n4. Risk Mitigation Recommendations:")
print("-" * 40)

recommendations = {
    "Urban Areas (Cluster 0)": [
        "Maintain high vaccination rates (Model Accuracy: 94.9%)",
        "Focus on public transport sanitization",
        "Implement dense-area monitoring systems",
        "Enhance air quality monitoring"
    ],
    "Rural Areas (Cluster 1)": [
        "Increase vaccination coverage (High Priority)",
        "Improve healthcare accessibility",
        "Deploy mobile health units",
        "Strengthen emergency response capabilities"
    ],
    "General Recommendations": [
        "Continuous monitoring with model validation",
        "Regular health infrastructure assessment",
        "Early warning system implementation",
        "Community health education programs"
    ]
}

for area, recs in recommendations.items():
    print(f"\n{area}:")
    for i, rec in enumerate(recs, 1):
        print(f"{i}. {rec}")

# Create implementation priority matrix
print("\n5. Implementation Priority Matrix:")
print("-" * 40)
priority_matrix = pd.DataFrame({
    'Region_Type': ['Rural', 'Urban', 'Suburban'] * 2,
    'Cluster': [1, 0, 1, 0, 1, 0],
    'Priority': ['High', 'Medium', 'Medium', 'Low', 'Medium', 'Low'],
    'Primary_Focus': [
        'Vaccination Coverage',
        'Dense Area Monitoring',
        'Healthcare Access',
        'Air Quality',
        'Mobile Healthcare',
        'Transport Safety'
    ],
    'Model_Performance': [f"{roc_auc:.1%}"] * 6
})
print(priority_matrix)

print("\n6. Monitoring Metrics:")
print("-" * 40)
print("Key Performance Indicators (KPIs):")
print("1. Vaccination Rate Changes")
print("2. Healthcare Accessibility Improvements")
print("3. Outbreak Response Time")
print("4. Environmental Risk Levels")
print("5. Public Health Engagement")
print(f"6. Model Performance (Current: {roc_auc:.1%})")

# Clean up
gc.collect()
```
